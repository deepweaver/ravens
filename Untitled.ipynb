```python
def crop(img, coor): 
    assert img.shape == (1400,1400,3) 
    results = [] 
    x = [] 
    y = [] 
    w = [] 
    h = [] 
    for i, v in enumerate(coor): 
        co = v[0] 
        assert len(co) == 4 
        if len(coor) == 10:
            if i >= 4:
                x.append(co[0]) 
                y.append(co[1]) 
                w.append(co[2]) 
                h.append(co[3])
        if len(coor) == 17: 
            if i >= 9:
                x.append(co[0]) 
                y.append(co[1]) 
                w.append(co[2]) 
                h.append(co[3])       
    print("w = ", len(w))
    a = 22 
    b = 130
    results.append(img[0:y[0]-b,:,:].copy()) 
    if len(coor) == 10:
        for i in range(6): 
            results.append(img[y[i]-a:y[i]+h[i]+a, x[i]-a:x[i]+w[i]+b,:].copy())  
    if len(coor) == 17: 
        for i in range(8): 
            results.append(img[y[i]-a:y[i]+h[i]+a, x[i]-a:x[i]+w[i]+b,:].copy())      
    print(len(results))
    return results
```

```python
def cut(img):
    # crop image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    th, threshed = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY_INV)

    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (11,11))
    morphed = cv2.morphologyEx(threshed, cv2.MORPH_CLOSE, kernel)

    _, cnts, _ = cv2.findContours(255-morphed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnt = sorted(cnts, key=cv2.contourArea)[-1]
    x,y,w,h = cv2.boundingRect(cnt)
    new_img = img[y:y+h, x:x+w]

    return new_img  
```

```python
def transBg(img):   
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    th, threshed = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY_INV)

    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (11,11))
    morphed = cv2.morphologyEx(threshed, cv2.MORPH_CLOSE, kernel)
#     print(morphed[:10,:10])
    morphed = morphed
    _, roi, _ = cv2.findContours(morphed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

    mask = np.zeros(img.shape, img.dtype)

    cv2.fillPoly(mask, roi, (255,)*img.shape[2], )

    masked_image = cv2.bitwise_and(img, mask)

    return masked_image
```

```python
def fourChannels(img):
    height, width, channels = img.shape
    if channels < 4:
        new_img = cv2.cvtColor(img, cv2.COLOR_BGR2BGRA)
        return new_img

    return img
```

```python
s_img = ret[3]

# set to 4 channels
s_img = fourChannels(s_img)

# remove white background
s_img = cut(s_img)

# set background transparent
s_img = transBg(s_img)
cv2.imwrite("img.png", s_img)
```

```python
def imgCroppedWrite(img, path): 
    s_img = img

    # set to 4 channels
    s_img = fourChannels(s_img)

    # remove white background
    s_img = cut(s_img)

    # set background transparent
    s_img = transBg(s_img)
    cv2.imwrite(path, s_img) 
```
